# ZOMATO DATA ANALYSIS PROJECT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Loading

```python
df= pd.read_csv("D:\Data Analytics Practice\Zomato_Dataset.csv")
# Displays first five rows
df.head()  
# Check Dataset Shape (rows and columns)
df.shape            
# Display column names to understand dataset structure
df.columns
df.info()
```

# Data Cleaning

```python
df.isnull().sum()
df.describe()
df. duplicated().sum()
df= df.drop_duplicates()
```

# Exploratory Data Analysis

```python
sns.countplot(x='Online_order', data= df)
plt.title ("Online Order Availability")
plt.show()
# Most Restaurants do not provide online Ordering
# This shows inconvenience to customers


sns.countplot(x='Book_table', data= df)
plt.title ("Table Booking Availability")
plt.show()


df['Rate']= df['Rate'].str.replace('/5','')
df['Rate']= df['Rate'].replace(['NEW','-'],None)
df['Rate']= pd.to_numeric(df['Rate'])
sns.histplot(df['Rate'], bins=10)
plt.title ("Rating Distribution")
plt.show()
# Most Ratings lie between 3.5 to 4.0
# Very few restaurants have extremely high or low ratings


## Relationship Analysis
sns.scatterplot(x= 'Approx_cost(for two people)', y= 'Votes', data= df)
plt.title("Cost vs Votes")
plt.show()
# Some low-cost restaurants have high votes
# Online availability may increase popularity


sns.boxplot(x= 'Online_order',y= 'Rate', data= df)
plt.title("Online Order vs Ratings")
plt.show()
# Restaurants that provide online ordering generally have slightly higher ratings 
# This shows customer prefer convenience

df.groupby('Online_order') ['Rate'].mean()
# Average rating is higher for restaurants that offer online ordering
```

# Insights

```python
# Final Conclusion:-

# Most Restaurants have ratings between 3.5- 4.0
# Online Ordering increases engagement 
# Cost does not always determine popularity 
# Table Booking is less common 
```
